```python
import networkx as nx
```

```python
#creating the element to merge networks
G = nx.DiGraph() 
```

```python
#working with TRRUST information
f = open("trrust_rawdata.human.tsv","r")
for line in f:
    line = line[:-1].split("\t")
    node1 = line[0].upper()
    node2 = line[1].upper()
    if G.has_edge(node1,node2): #previously I have added the edge, so I will look ig the effect is the same, if not, I will append the new one
        if line[2] != "Unknown":
            if G[node1][node2]["effect"] != line[2]:
                if "Unknown" not in G[node1][node2]["effect"]:
                    G[node1][node2]["effect"] = G[node1][node2]["effect"]+", "+line[2]           
                else:
                    if line[2] not in G[node1][node2]["effect"]:
                        G[node1][node2]["effect"] = line[2]
    else:
        G.add_edge(node1, node2, effect=line[2], found_in="TRRUST")
f.close()    
```

```python
#working with DoRothEA
f = open("dorothea_human_1.4.1_ABC.tsv","r")

for line in f:
    node1,node2 = line[:-1].split("\t")
    node1 = node1.upper()
    node2 = node2.upper()
    if G.has_edge(node1,node2):
        G[node1][node2]["found_in"] = G[node1][node2]["found_in"]+", DoRothEA"
    else:
        G.add_edge(node1, node2, effect="Unknown", found_in="DoRothEA")
f.close()
```

```python
#working with RegNetwork
f = open("export_Tue_Dec_14_22_21_08_UTC_2021.csv","r")
for line in f:
    line = line[:-1].split(",")
    if line[-1] == "High":
        node1 = line[0].upper()
        node2 = line[2].upper()
        if G.has_edge(node1,node2):
            G[node1][node2]["found_in"] = G[node1][node2]["found_in"]+", RegNetwork"
        else:
            G.add_edge(node1, node2, effect="Unknown", found_in="RegNetwork")

f.close()
```

```python
for edge in G.edges(data=True):
    print(edge)
```

```python
#saving net
nx.write_gml(G,"merge_trrust_dorothea_regnetwork.gml")
```
